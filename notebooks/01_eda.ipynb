```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# Load Tuesday's data (FTP & SSH Brute Force attacks)
df = pd.read_csv('../data/MachineLearningCVE/Tuesday-WorkingHours.pcap_ISCX.csv')

# Basic exploration
print("Shape:", df.shape)
print("\nColumn names:", df.columns.tolist())
print("\nClass distribution:")
print(df[' Label'].value_counts())
```

```python
# Visualize class distribution
plt.figure(figsize=(8, 5))
colors = ['steelblue', 'crimson', 'darkorange']
df[' Label'].value_counts().plot(kind='bar', color=colors)
plt.title('Attack Type Distribution - Tuesday Dataset', fontsize=14)
plt.xlabel('Traffic Type')
plt.ylabel('Number of Connections')
plt.xticks(rotation=0)
plt.tight_layout()
plt.savefig('../outputs/class_distribution.png')
plt.show()
print("Chart saved to outputs folder!")
```

```python
# Check for missing or infinite values
print("Missing values:", df.isnull().sum().sum())
print("Infinite values:", np.isinf(df.select_dtypes(include=np.number)).sum().sum())
print("\nData types:")
print(df.dtypes.value_counts())
print("\nBasic statistics:")
df.describe()
```

```python
df.replace([np.inf,-np.inf], np.nan, inplace=True)

df.dropna(inplace=True)

print("After cleaning:")
print("Shape:", df.shape)
print("Missing values:", df.isnull().sum().sum())
print("Infinite values:", np.isinf(df.select_dtypes(include=np.number)).sum().sum())
```

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler

# Separate features (X) and target (y)
X = df.drop(' Label', axis=1)  # everything except the label
y = df[' Label']               # just the label column

# Convert text labels to numbers
# BENIGN=0, FTP-Patator=1, SSH-Patator=2
le = LabelEncoder()
y_encoded = le.fit_transform(y)

print("Classes found:", le.classes_)
print("Encoded as:", list(range(len(le.classes_))))
print("\nFeature shape:", X.shape)
print("Label shape:", y_encoded.shape)
```

```python
from sklearn.model_selection import train_test_split

# Split into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(
    X, y_encoded, 
    test_size=0.2,      # 20% for testing, 80% for training
    random_state=42,    # makes results reproducible
    stratify=y_encoded  # keeps class balance the same in both sets
)

print("Training set size:", X_train.shape)
print("Testing set size:", X_test.shape)
print("\nTraining class distribution:")
print(pd.Series(y_train).value_counts())
print("\nTesting class distribution:")
print(pd.Series(y_test).value_counts())
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

print("Data scaled successfully!")
print("Mean of first feature (should be ~0):", X_train_scaled[:, 0].mean().round(4))
print("Std of first feature (should be ~1):", X_train_scaled[:, 0].std().round(4))
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix
import time

# Train the Random Forest model
print("Training Random Forest... (this may take a minute)")
start_time = time.time()

rf_model = RandomForestClassifier(
    n_estimators=100,    # 100 decision trees
    random_state=42,     # reproducible results
    n_jobs=-1            # use all CPU cores to speed up training
)

rf_model.fit(X_train_scaled, y_train)

end_time = time.time()
print(f"Training complete! Time taken: {round(end_time - start_time, 2)} seconds")

# Test the model
y_pred = rf_model.predict(X_test_scaled)

# Print results
print("\n--- Model Performance ---")
print(classification_report(y_test, y_pred, target_names=le.classes_))
```

```python
# Plot confusion matrix
plt.figure(figsize=(8, 6))
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=le.classes_,
            yticklabels=le.classes_)
plt.title('Confusion Matrix - Random Forest', fontsize=14)
plt.ylabel('Actual Label')
plt.xlabel('Predicted Label')
plt.tight_layout()
plt.savefig('../outputs/confusion_matrix.png')
plt.show()
```

```python
import shap

print("Calculating SHAP values... (this may take a few minutes)")

# Create SHAP explainer using a sample of data (full dataset is too slow)
sample_size = 500
X_sample = pd.DataFrame(X_test_scaled[:sample_size], columns=X.columns)

explainer = shap.TreeExplainer(rf_model)
shap_values = explainer.shap_values(X_sample)

print("SHAP values calculated!")
```

```python
# Summary plot - shows most important features
shap.summary_plot(
    shap_values[:, :, 1],  # class 1 = FTP-Patator
    X_sample,
    plot_type="dot",
    max_display=15,        # show top 15 features
    show=False
)
plt.title("Top 15 Most Important Features for FTP Attack Detection")
plt.tight_layout()
plt.savefig('../outputs/shap_summary.png')
plt.show()
```
